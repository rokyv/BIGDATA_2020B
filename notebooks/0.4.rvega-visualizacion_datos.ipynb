```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Diagramas de Dispersion (scatterplot)

x = np.arange(2,20)
y = x**2
y2 = x**3

plt.scatter(x,y, color="black", marker="*", label="x**2")
plt.scatter(x,y2, color="green", marker="s", label="x**3")
plt.plot(x, y2, color="pink", label="line")


plt.title("Funciones Polinomicas", fontsize = 20, color="red")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()

plt.text(5,4000,"texto", color="blue")  #se ecrbien las coordenadas
```

```python
# Histogramas

mean = 5
std = 3
n = 10000
#x = np.random.normal(5, 3, 100)
x = np.random.normal(mean, std, n) #genera numeros aleatorios al rededor, que la desviaci칩n sea esto, y que coloque cuantos numeros
plt.hist(x, bins=20); #el ; al final es para que no salgan los datos
# cada "columna" (caja) o "bean" es de la distribuci칩n de los datos cuantos caen en cada caja

plt.title("Funciones de una variable aleatoria", fontsize = 20, color="red")
plt.xlabel("x")
plt.ylabel("frecuencia")
plt.grid()
plt.legend()

plt.text(4,1400,f"mean: {mean} +/- 3", color="blue")  #se ecrbien las coordenadas

```

```python
plt.boxplot(x, vert=True) # en los extremos muestra los outlyers o datos atipico. en el centro est치n los cuartiles el valor medio es el cuartol침 2 (50%)

```

```python
# Subplots o subgraficos.
# en donde se pinta es un lienzo. y se puede dividir o ajustar para mostrar varias graficas

fig, ax = plt.subplots(ncols=2, nrows=1, figsize=(15,5)) #cuantas columnas a la izquierda y a la derecha

ax[0].hist(x);
ax[1].boxplot(x, vert=False);



```

```python
fig, ax = plt.subplots(ncols=1, nrows=2, figsize=(5,10), sharex= True) 

ax[0].hist(x);
ax[1].boxplot(x, vert=False);
```

```python
# Barplot

classes = ["gatos", "perros", "ratones"]
totales = [20,35,12]

plt.bar(classes, totales)   #bar es vertical y barh es horizontal
```

```python
import seaborn as sns 
# se requiere un dataframe para construir

data = pd.DataFrame({
    "Animales": classes,
    "Totales": totales
    
})

data.head()

# sns.barplot(x="Animales", y="Totales", data = data)

plots = sns.barplot(x="Animales", y="Totales", data = data)

for bar in plots.patches:
    plots.annotate(format(bar.get_height(),".1f"),
                   (bar.get_x() + bar.get_width() /2,
                    bar.get)
                   
                   )
```
